# Slices an image into equal subparts

```python
from matplotlib.image import imread
from urllib.request import urlopen
import matplotlib.pyplot as plt
from scipy.misc import imread
from scipy.misc import imsave
import os
```

```python
path_save = ''
name = 'Test'
file_exctension = 'png'

url = 'https://upload.wikimedia.org/wikipedia/de/f/f4/Braess_example_times.jpg'

with urlopen(url) as file:
    image = imread(file, mode='RGB')
```

```python
h, b, t = image.shape
```

```python
f, axarr = plt.subplots(2, 2)
axarr[0,0].imshow(image[:int(h/2), :int(b/2)])
axarr[0,1].imshow(image[:int(h/2), int(b/2):])
axarr[1,0].imshow(image[int(h/2):, :int(b/2)])
axarr[1,1].imshow(image[int(h/2):, int(b/2):])
plt.show()
```

```python
h, b, t = image.shape
    
imsave(path_save+name+'_1.'+file_exctension, image[:int(h/2), :int(b/2)])
imsave(path_save+name+'_2.'+file_exctension, image[:int(h/2), int(b/2):])
imsave(path_save+name+'_3.'+file_exctension, image[int(h/2):, :int(b/2)])
imsave(path_save+name+'_4.'+file_exctension, image[int(h/2):, int(b/2):])
print('Finished')
```
